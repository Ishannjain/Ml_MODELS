## Python String
- Python string is the collection of the characters surrounded by single quotes, double quotes, or triple quotes. 
- The computer does not understand the characters; internally, it stores manipulated character as the combination of the 0's and 1's.

- Each character is encoded in the ASCII or Unicode character. 
- So we can say that Python strings are also called the collection of Unicode characters.

```python
str="hello"
print(str,"=", type(str))
str='hello'
print(str,"=", type(str))
str='''hello'''
print(str,"=", type(str))
print(str[0])
print(ascii(str[0]))
print(ord(str[0]))
#Using triple quotes  
str3 = '''''Triple quotes are generally used for  
    represent the multiline or 
    docstring'''  
print(str3)
str4="kps harma \
classes"
print(str4)
```

## Strings indexing and splitting
- Index of string starts from 0 to length-1
- Both positive and negative index are there
- The slice operator [] is used to access the individual characters of the string. 
- However, we can use the : (colon) operator in Python to access the substring from the given string.


```python
str="Python-String"
print(str[-13])
print(str[0])
print(str[:])
print(str[0:])
print(str[4:])
print(str[:5])
print(str[-12:-2])
print(str[11:4])
print(str[-2:-12])
print(str[-12:-2])
```

## String Operators
<p>
1. <b>+</b>	It is known as concatenation operator used to join the strings given either side of the operator.<br>
2. <b>*</b>	It is known as repetition operator. It concatenates the multiple copies of the same string. <br>
3. <b>[]</b>	It is known as slice operator. It is used to access the sub-strings of a particular string.<br>
4. <b>[:]</b>	It is known as range slice operator. It is used to access the characters from the specified range.<br>
5. <b>in</b>   It is known as membership operator. It returns if a particular sub-string is present in the specified string.<br>
6. <b>not in</b>	It is also a membership operator and does the exact reverse of in. It returns true if a particular substring is not present in the specified string.<br>
7. <b>r/R</b>	It is used to specify the raw string. Raw strings are used in the cases where we need to print the actual meaning of escape characters such as "C://python". To define any string as a raw string, the character r or R is followed by the string.<br>
8. <b>%</b>	It is used to perform string formatting. It makes use of the format specifiers used in C programming like %d or %f to map their values in python. We will discuss how formatting is done in python.</p>

```python
str = "Hello"     
str1 = " world"    
print(str*3) # prints HelloHelloHello    
print(str+str1)# prints Hello world     
print(str[4]) # prints o                
print(str[2:4]); # prints ll                    
print('w' in str) # prints false as w is not present in str    
print('wo' not in str1) # prints false as wo is present in str1.     
print(r'C://python\t37') # prints C://python37 as it is written    
print("The string str : %s this is %s string"%(str, str1)) # prints The string str : Hello  
```

### isalnum()
- isalnum() method checks whether the all characters of the string is alphanumeric or not. A character which is either a letter or a number is known as alphanumeric. It does not allow special chars even spaces.

### isalpha() 
- It returns true if all the characters are alphabets and there is at least one character, otherwise False.

### isdecimal()	
- It returns true if all the characters of the string are decimals.

### isdigit()	
- Python isdigit() method returns True if all the characters in the string are digits. It returns False if no character is digit in the string.

### isidentifier()	
- It returns true if the string is the valid identifier.

### islower()	
- It returns true if the characters of a string are in lower case, otherwise false.

### isnumeric()	
- It returns true if the string contains only numeric characters.

### isprintable()	
- It returns true if all the characters of s are printable or s is empty, false otherwise.

### isupper()
- It returns false if characters of a string are in Upper case, otherwise False.

### isspace()	
- isspace() method is used to check space in the string. 
- It return true if there are only whitespace characters in the string. Otherwise it returns false. 
- Space, newline, and tabs etc are known as whitespace characters

### istitle()	
- It returns true if the string is titled properly and false otherwise. A title string is the one in which the first character is upper-case whereas the other characters are lower-case.


```python
str1="hello123"
str2="hello123#"
str3="hello"
print(str1.isalnum())
print(str2.isalnum())
print(str3.isalpha())
str4="545"
print(str4.isdecimal())
str4="0"
print(str4.isdigit())
```

```python
c = '\u00B2'
c1='\u00BD'
print(c)
print(c1)
print(c.isdecimal())
print(c.isdigit())
print(c.isnumeric())
print(c1.isdecimal())
print(c1.isdigit())
print(c1.isnumeric())
```

### ljust(width[,fillchar])
- This method left justify the string and fill the remaining spaces with fillchars. 
- It returns a new string justified left and filled with fillchars.

### rjust(width,[,fillchar])
- Returns a space padded string having original string right justified to the number of characters specified.

### zfill(width)
- Returns original string leftpadded with zeros to a total of width characters; intended for numbers

### lstrip([chars])
- lstrip() method is used to remove all leading characters from the string. 
- It takes a char type parameter which is optional. If parameter is not provided, it removes all the leading spaces from the string.

### rstrip()
- It removes all trailing whitespace of a string and can also be used to remove particular character from trailing.

### strip([chars])
- It is used to perform lstrip() and rstrip() on the string.

### partition()
- It splits the string from the string specified in parameter. 
- It splits the string from at the first occurrence of parameter and returns a tuple. 
- The tuple contains the three parts before the separator, the separator itself, and the part after the separator.
- It returns an empty tuple having seperator only, if the seperator not found.

### replace(old,new[,count])
- Return a copy of the string with all occurrences of substring old replaced by new. 
- If the optional argument count is given, only the first count occurrences are replaced.

### split(str,num=string.count(str))
- Splits the string according to the delimiter str. The string splits according to the space if the delimiter is not provided.
- Splits the string into a comma separated list. 

### rsplit(sep=None, maxsplit )
- It is same as split() but it processes the string from the backward direction. It returns the list of words in the string. 
- maxsplit: It is a number, which tells us to split the string into a maximum of provided number of times. If it is not provided then there is no limit.  
- If Separator is not specified then the string splits according to the white-space.

### splitlines(num=string.count('\n'))
- It returns the list of strings at each line with newline removed.
- Splits the string based on the lines. It breaks the string at line boundaries and returns a list of splitted strings. 
- Line breakers can be a new line (\n), carriage return (\r), paragraph end, etc.
