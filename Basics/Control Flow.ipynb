### If..else statement syntax
 <p> if condition:<br>
      &nbsp;&nbsp;&nbsp;statement1 <br>
      else: <br>
      &nbsp;&nbsp;&nbsp; statement2 <br>
      statement2 </p>

```python
a=2
if(a==2):
    print("a is 2")
else: 
    print("a is not 2")
print("Next Line")
```

## Nested if else
<p> if (condition1): <br>
   &nbsp;&nbsp;&nbsp; Executes when condition1 is true  <br>
   &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if (condition2): <br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Executes when condition2 is true <br>
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if Block is end here <br>
# if Block is end here </p>

```python
x,y=2,3
if x==2:
    print("inside outer if")
    if(y==2):
        print("inside inner if")
    print("exit outer if")
print("after outer if")
    
```

```python
x,y=2,3
if x==2:
    print("inside outer if")
    if(y==2):
        print("inside inner if")
    else:
        print("inside else")
    print("exit outer if")
print("after outer if")
```

## if--elif ladder
<p>if condition1:<br>
    &nbsp;&nbsp;&nbsp; statement1 <br>
    elif condition2: <br>
    &nbsp;&nbsp;&nbsp; statement2 <br>
    elif condition3: <br>
    &nbsp;&nbsp;&nbsp; statement3 <br>
    elif condition4: <br>
    &nbsp;&nbsp;&nbsp; statement4 <br>
    else: <br>
    statement5 </p>




```python
x=int(input("input a number for printing day"))
if x==1:
    print("Sun")
elif x==2:
    print("Mon")
elif x==3:
    print("Tue")
elif x==4:
    print("Wed")
elif x==5:
    print("Thu")
elif x==6:
    print("Fri")
elif x==7:
    print("Sat")
else:
    print("No Day Defined")
```

## for loop or for in loop
<i> <b>Syntax</b> </i> <br>
for <variable> in range(start_val, end_val, incr) <br>

```python
a=int(input("Enter a number"))
for i in range(a):
    print(i,end="\t")
```

```python
# for loop with list

print("List Iteration")
l = ["ram", "shyam", "kp","Ap"]
for i in l:
    print(i,end="\t")
      
# Iterating over a tuple (immutable)

print("\nTuple Iteration")
t = ("ram", "shyam", "kp","Ap")
for i in t:
    print(i,end="\t")
      
# Iterating over a String

print("\nString")   
s = "Python"
for i in s :
    print(i,end="\t")
      
# Iterating over dictionary
print("\nDictionary Iteration")  
d = dict()
d['xyz'] = 123
d['abc'] = 345
for i in d :
    print("%s  %d" %(i, d[i]),end="\t")
```

## Nested Loop


```python
a=int(input("Enter Number of rows"))
b=int(input("Enter Number of columns"))
for i in range(a):
    for j in range(b):
        print("*", end=" ")
    print("\n")
```

```python
a=int(input("Enter Number of rows"))
b=int(input("Enter Number of columns"))
for i in range(a):
    for j in range(b):
        if j<=i:
            print("*", end=" ")
    print("\n")
```

```python
a=int(input("Enter Number of rows"))
b=int(input("Enter Number of columns"))
for i in range(a):
    for j in range(b):
        if j<i:
            print(" ", end=" ")
    for j in range(b):
        if j<b-i:
            print("*", end=" ")
    print("\n")
```

## Loop Control Statements
- There are three such statements in python
1. break
2. continue
3. pass



```python
for i in range(10):
    if(i==5):
        break
    print(i,end=" ")
```

```python
for i in range(10):
    if(i==5):
        continue
        i=i+6
    print(i,end=" ")
```

```python
for i in range(10):
    if(i==5):
        pass
        i=i+6
    print(i,end=" ")
```

```python
for i in range(4):
    for j in range(4):
        if(i==j):
            break
        print(i," ",j)
```

```python
# Program for printing prime numbers from some range
a=1#int(input("Enter starting number"))
b=100#int(input("Enter Ending number"))
for i in range(a,b+1):
    flag=True
    if i>1:
        for j in range(2,int(i/2)+1):
            if(i%j==0):
                flag=False
                break
        if flag==True: 
            print(i,end="\t")
```
