<font color=Blue><h2> List </h2> </font> 
- Lists are similar to arrays in C but here list can contain data of different types. 
- The items stored in the list are separated with a comma (,) and enclosed within square brackets [].
- We can use slice [:] operators to access the data of the list. 
- The concatenation operator (+) and repetition operator (*) works with the list in the same way as they were working with the strings.

```python
list1  = [1, "hi", "Python", 2, 3.5]    
#Checking type of given list  
print(type(list1))  

#Printing the list1  
print (list1)  

print(list1[2])
# List slicing  

print (list1[3:])  
  
# List slicing  
print (list1[0:2])   
  
# List Concatenation using + operator  
print (list1 + list1)  
  
# List repetation using * operator  
print (list1 * 3) 

# Updating the list
list1[1] = 5.5 

print(list1)
print(list1[-1])
```

<font color=Blue><h2> Tuples </h2> </font> 
- Like lists, tuples also contain the collection of the items of different data types. 
- The items of the tuple are separated with a comma (,) and enclosed in parentheses ().

```python
tup  = ("hi", "Python", 2)    
# Checking type of tup  
print (type(tup))    
  
#Printing the tuple  
print (tup)  
  
# Tuple slicing  
print (tup[1:])    
print (tup[0:1])    
  
# Tuple concatenation using + operator  
print (tup + tup)    
  
# Tuple repatation using * operator  
print (tup * 3)  
print(tup)
  
# Adding value to tup. It will throw an error.  
#tup[2] = "hi"  
```

<font color=Blue><h2> Dictionary </h2> </font> 
- Dictionary is an unordered set of a key-value pair of items. 
- It maintains the data elements in key value pairs separated by , in {} braces <b>ex {key1:value1, key2:value2,... } </b>
- It is like a hash table where each key stores a specific value. 
- Key can hold any primitive data type, whereas value is an arbitrary Python object.
- Change is allowed here


```python
d = {1:'kp', 2:6.5, 3:'shyam', 4:'mohan'}     
  
# Printing dictionary  
print (d)  
  
# Accesing value using keys  
print("1st name is "+d[1])   
print("2nd name is "+ d[4]) 

d[2]='ram'

print(d)
print (d.keys())    
print (d.values())    
```

<font color=Blue><h2> Set </h2> </font> 
- Set is the unordered collection of the data type. 
- It is iterable, mutable(can modify after creation), and has unique elements. 
- In set, the order of the elements is undefined; it may return the changed sequence of the element. 
- The set is created by using a built-in function set(), or a sequence of elements is passed in the curly braces and separated by the comma. 
- It can contain various types of values

```python
# Python program to demonstrate 
# Creation of Set in Python

set5={1,2,3,4,5}
print(set5)
# Creating an empyty Set

set1 = set()
print("Initial blank Set: ")
print(set1)
   
# Creating a Set with  String
set1 = set("KP Sharma")
print(set1)
 
# Creating a Set with List

set1 = set(["KP", "AP", "PK"])
print("\nSet with the use of List: ")
print(set1)
 
# Creating a Set with a mixed type of values

set1 = set([1, 2, 'kk', 4, 'For', 6, 'pp'])
print("\nSet with the use of Mixed Values")
print(set1)
```

### flush Argument
<p>The I/Os in python is generally buffered, meaning they are used in chunks. This is where flush comes in as it helps users to decide if they need the written content to be buffered or not. By default, it is set to false. If it is set to true, the output will be written as a sequence of characters one after the other.</p>

```python
import time
count = 4
for i in range(4):
    if i < 3:
        print(i, end='>>>')
        time.sleep(1)
    else:
        print('Start')
```

```python
import time
 
count = 4
for i in range(4):
    if i < 3:
        print(i, end='>>>', flush=False)
        time.sleep(1)
    else:
        print('Start')
```

## Python range() Function
- The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

### Syntax:  range(start, stop, step)
#### Parameter	Description
- start	Optional. An integer number specifying at which position to start. Default is 0
- stop	Required. An integer number specifying at which position to stop (not included).
- step	Optional. An integer number specifying the incrementation. Default is 1

```python
x=range(10,20)
print(x)
print(type(x))
```

```python
x = range(3, 9)
for n in x:
    print(n)
```

```python
x = range(3, 20, 2)
for n in x:
    print(n)
print(n)
```
