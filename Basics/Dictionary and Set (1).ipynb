<font color=Red><h1 align="center">  <strong> Dictionaries </strong> </h1> </font>

- Dictionaries are Python’s implementation of a data structure that is more generally known as an associative array. A dictionary consists of a collection of key-value pairs. Each key-value pair maps the key to its associated value.

- You can define a dictionary by enclosing a comma-separated list of key-value pairs in curly braces ({}). A colon (:) separates each key from its associated value.

- Key and Values in a dictionary can be of any datatype. The value can be duplicated, whereas keys can’t be repeated and must be immutable. 

- Dictionary is a mutable object so all elements of the dictionary can be changed, updated, or removed. 

- The dictionary takes O(1) look-up speed.

- Dictionary keys are case sensitive, same name but different cases of Key will be treated distinctly.

## Different ways of Creating a Dictionary
- The dictionary can be created by following ways

### By using { } braces

```python
dict1={}  # empty dictionary
print(dict1)

dict2= {'a':123, 'kp':234.56, 1:"ram"}  # dictionary with multiple key value pair
print(dict2)

dict3={10}   # if only one value is passed then it become set
print(type(dict3))

# Multiple elements can be added after creating a dictionary
dict1['kp']=100
dict1[1]="ram"
dict1[10.2]=234

print(dict1)

dict1[1]="Shyam"  # it will update the value for key 1

print(dict1)

dict2= {'a':123, 'a':234.56, 1:"ram"}  # Same keys are not allowed only last value is preserved in this case
print(dict2)

dict3={'a':534,'l1':[2,3,4],'tup':(3,4,5,6), 'dict':{1:"kp",2:"ap"}}  # can have other list as value of key
print(dict3)
```

## By the built-in function dict()

- Python dict() function is a constructor which creates a dictionary. Python dictionary provides three different constructors to a create dictionary.

- If no argument is passed, it creates an empty dictionary.

- If a positional argument is given, a dictionary is created with the same key-value pairs. Otherwise, pass an iterable object.

- If keyword (named) arguments are given, the keyword arguments and their values are added to the dictionary created from the positional argument.

### Syntax:
- dict ([**kwargs])  
- dict ([mapping, **kwargs])  
- dict ([iterable, **kwargs])  

### Parameters
- kwargs: It is a keyword argument.

- mapping: It is another dictionary.

- iterable: It should be an iterable object in the form of a key-value pair(s).


```python

# Calling function  
result = dict() # returns an empty dictionary  
result2 = dict(a=1,b=2)  
# Displaying result  
print(result)  
print(result2) 
```

```python
result = dict({'x': 5, 'y': 10}, z=20) # Creating dictionary using mapping  
result2 = dict({'x': 5, 'y': 10, 'z':20})  
# Displaying result  
print(result)  
print(result2)  
```

```python
# initializing the list
list_of_tuples= [('Key 1', 1), ('Key 2', 2), ('Key 3', 3), ('Key 4', 4), ('Key 5', 5)]

# converting to dict
result = dict(list_of_tuples)
tup1=([1,2],["kp",1])
print(dict(tup1))
# printing the result
print(result)

tuple1=((1,2),(3,4),[6,7])

print(dict(tuple1))
```

## Some important dictionary methods

### dict.clear()
- Removes all elements of dictionary dict

### dict.copy()
- Returns a shallow copy of dictionary dict

### dict.fromkeys(seq, value)
- Create a new dictionary with keys from seq and values set to value.
 - seq − This is the list of values which would be used for dictionary keys preparation.
 - value − This is optional, if provided then value would be set to this value

### dict.get(key, default=None)
- The method is used to get value by key
- If key not available then the message or value set to default will be returned.


###  dict.items()
- Returns a list of dict's (key, value) tuple pairs

###  dict.keys()
- Returns list of dictionary dict's keys

###  dict.setdefault(key, default=None)
- The setdefault() method returns the value of the item with the specified key.
- If the key does not exist, insert the key, with the specified value.
- key (Required)-The keyname of the item you want to return the value from
- value	(Optional) If the key exist, this parameter has no effect.
- If the key does not exist, this value becomes the key's value
- Default value None

###  dict.update(dict2)
- Adds dictionary dict2's key-values pairs to dict

###  dict.values()
- Returns list of dictionary dict's values

### dict.pop(key, Msg)
- The pop() method removes the specified item from the dictionary.
- The value of the removed item is the return value of the pop() method
- key(Required) The keyname of the item you want to remove
- Msg(Optional) A value to return if the specified key do not exist. 
- If specified key is not exist and you have not passed Msg then exception will be raised

<font color=Red><h1 align="center">  <strong> Set </strong> </h1> </font>

- A Set is an unordered collection data type that is iterable, mutable and has no duplicate elements. 
- Python’s set class represents the mathematical notion of a set. 
- Each element in the set must be unique, immutable, and the sets remove the duplicate elements. 
- Sets are mutable which means we can modify it after its creation.

### set() method 
- It is used to convert any of the iterable to sequence of iterable elements with distinct elements, commonly called Set.
- Syntax : set(iterable)
- Parameters : Any iterable sequence like list, tuple or dictionary.
- Returns : An empty set if no element is passed. Non-repeating element iterable modified as passed as argument. 

```python
s1=set("kp sharma")
print(s1)

tup=(3,4,5,6,7,8)
s2=set(tup)
print(s2)

list1=[3,4,5,6,7,6]
print(set(list1))

dict1={"kp":2425,"ap":784}
s3=set(dict1)
print(s3)
```

## Methods for Sets
### Adding elements
- Insertion in set is done through set.add() function, where an appropriate record value is created to store in the hash table.

```python

people = {"Jay", "Idrish", "Archi"}
  
print(people)
  
# This will add Daxit
# in the set
people.add("Daxit")
  
# Adding elements to the
# set using iterator
for i in range(1, 6): 
    people.add(i) 
  
print("\nSet after adding element:", end = " ")
print(people)
```

```python
Months = set(["January","February", "March", "April", "May", "June"])    
  
print(Months)    
   
Months.add("July");    
Months.add ("August");    
  
print(Months)  

print("\nlooping through the set elements ... ")    
for i in Months:    
    print(i)    
```

###  update() method 
- To add more than one item in the set,. It accepts iterable as an argument.


```python
Months = {"January","February", "March", "April", "May", "June"}    
  
print(Months)    
set2= {"July","August","September","October"}  
Months.update(set2);    

print("\nprinting the modified set ... ")     
print(Months); 
```

### Removing items from the set
## discard(item) & remove(item) methods
- discard() method and remove() method are used to remove the items from the set. 
- The difference between these function, using discard() function if the item does not exist in the set then the set remain unchanged whereas remove() method will through an error.

```python
months = set(["January","February", "March", "April", "May", "June"])    
 
print(months)  

print("\nRemoving some months from the set...");    
months.discard("January");    
months.discard("May");  
print(months)  

months.discard("May"); 
print("\nPrinting the modified set...");    
print(months)    
  
```

```python
months = set(["January","February", "March", "April", "May", "June"])    
   
print(months)

print("\nRemoving some months from the set...");    
months.remove("January");    
months.remove("May");  
print(months) 
months.remove("May"); # will throw KeyError error
print("\nPrinting the modified set...");    
print(months)  
```

###  pop() method 
- to remove the item. Generally, the pop() method will always remove the last item but the set is unordered, we can't determine which element will be popped from set.

```python
Months = set(["January","February", "March", "April", "May", "June"])    
   
print(Months)  

print("\nRemoving some months from the set...");    
Months.pop();    
Months.pop();    

print("\nPrinting the modified set...");    
print(Months)
```

## Set Operations

### Union Operation
- Two sets can be merged using union() function or | operator. 
- The duplicate items are removed. 
- Time Complexity of this is O(len(s1) + len(s2)) where s1 and s2 are two sets whose union needs to be done

```python
s1={1,2,3,4}
s2={2,5,6,7,8,4}
s3=s1.union(s2)
print(s3)

s3=s1|s2
print(s3)
```

### Intersection
- This can be done through intersection() or & operator. Common Elements are selected. .

```python
s1={1,2,3,4}
s2={2,5,6,7,8,4}
s3=s1.intersection(s2)
print(s3)

s3=s1&s2
print(s3)
```

### Difference
- To find difference in between sets. This is done through difference() or – operator. 

```python
s1={1,2,3,4}
s2={2,5,6,7,8,4}
s3=s1.difference(s2)
print(s3)

s3=s1-s2
print(s3)
```

### Clear() method 
- empties the whole set.

```python
set1 = {1,2,3,4,5,6}
  

print(set1)
  

set1.clear()
  
print(set1)
```

## difference_update(....)	
- It modifies this set by removing all the items that are also present in the specified sets.
- Syntax:: set.difference_update(set1,set2,...)
- Returns None

```python
s1={1,2,3,4}
s2={2,5,6,7,8,4}
s4={3,4,5}
s3=s2.difference_update(s1,s4)
print(s1)
print(s2)
print(s3)
```

## intersection_update(....)	
- It removes the items from the original set that are <b>not present</b> in both the sets (all the sets if more than one are specified).
- Syntax:: set.intersection_update(set1,set2,...)
- Returns None

```python
s1={1,2,3,4}
s2={2,5,6,7,8,4}
s4={3,4,5}
s3=s1.intersection_update(s2)
print(s1)
print(s2)
print(s3)
```
